###`Q-5:` In IPL matches dataset some teams name has changed.
You will have to consider them as same.
```
'Delhi Capitals' formerly as 'Delhi Daredevils'
'Punjab Kings' formerly as 'Kings XI Punjab'
'Rising Pune Supergiant' formerly as 'Rising Pune Supergiants'
```
You need to make changes accordingly. Consider current name for each teams.

Be careful Gujrat Titans and Gujrat Lions are different teams.

````python
import os
cwd = os.getcwd()
ipl_df = pd.read_csv(os.path.join(cwd, "IPL_Matches_2008_2022.csv"))
# poc for one team
ipl_df[ipl_df[['Team1', 'Team2', 'TossWinner', 'WinningTeam']] == 'Delhi Daredevils'] = "Delhi Capitals"

(ipl_df[['Team1', 'Team2', 'TossWinner', 'WinningTeam']] == 'Delhi Daredevils').shape

# actual implementation
cols = ['Team1', 'Team2', 'TossWinner', 'WinningTeam']

team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}

ipl_df[cols] = ipl_df[cols].replace(team_mapping) # using replace


# after
(ipl_df == 'Delhi Daredevils').sum().sum()
````

###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
import matplotlib.pyplot as plt
def team1_vs_team2(team1, team2):
    mask = ((ipl_df['Team1'] == team1) & (ipl_df['Team2'] == team2)) | \
        ((ipl_df['Team1'] == team2) & (ipl_df['Team2'] == team1))
    
    matches = ipl_df[mask]

    team1_wins = (matches['WinningTeam'] == team1).sum()
    team2_wins = (matches['WinningTeam'] == team2).sum()

    top_players = matches['Player_of_Match'].value_counts().head(5)

     # ---- Plotting ----
    plt.figure(figsize=(12,5))

    # 1. Win-Loss Chart
    plt.subplot(1,2,1)
    plt.bar([team1, team2], [team1_wins, team2_wins])
    plt.title("Win Comparison")

    # 2. Player of Match
    plt.subplot(1,2,2)
    top_players.plot(kind='bar')
    plt.title("Top Players of Match")

    plt.show()

    # ---- Summary ----
    print("Total Matches:", len(matches))
    print(f"{team1} Wins:", team1_wins)
    print(f"{team2} Wins:", team2_wins)

    print("Top Player:", top_players.idxmax())

team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
````
