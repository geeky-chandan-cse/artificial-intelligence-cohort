###`Q-1:` Find the nearest element in the array to a given integer.
```
Ex:-
a=23 and array - [10 17 24 31 38 45 52 59].
Nearest element is 24
```
Hint: Read about this function `argmin()`


````python
arr = np.array([10, 17, 24 ,31 ,38 ,45 ,52 ,59, 22])
a = 23

diffrences_arr = np.abs(arr - a) # find the diffrence b/w each element with a
nearest_element = np.argmin(diffrences_arr) # find the min element index, bcz that will be the element with lowest value 
print("nearest element", arr[nearest_element])


print(diffrences_arr, nearest_element)

# find all nearest element 
min_num = np.min(diffrences_arr)
indices = np.where(diffrences_arr == min_num)

indices
````

###`Q-2:` Replace multiples of 3 or 5 as 0 in the given array.
```
arr=[1 2 3 4 5 6 7 9]

result-> [1 2 0 4 0 0 7 0]
```

###`Q-4:` Your are given an array which is havig some nan value. You job is to fill those nan values with most common element in the array.
```
arr=np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])

```

````python
arr = np.array([[1,2,np.nan],[4,2,6],[np.nan,np.nan,5]])
# 1. Find bool_arr of nan values
bool_arr_nan = np.isnan(arr)
print("nan indices", bool_arr_nan)

# 2. Filter array and remove np.nan values
# filtered_arr = arr[~bool_arr_nan]

# 3. Find most common/repeated element
# values, counts = np.unique(filtered_arr, return_counts=True)

# 2 and 3 combined to save extra spaces
values, counts = np.unique(arr[~bool_arr_nan], return_counts=True)
most_common = values[np.argmax(counts)]
print("array before upate", arr)
arr[bool_arr_nan] = most_common
print("array after update", arr)
````

###`Q-7:` Plot this below equation.

$$\large y = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

**Note: This equation is called tanh activation function. In deep learning, many times this function is used. If you find some difference between the sigmoid function and this tanh function, note that to your notebook.**

````python
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 100)
exp_x_positve = np.exp(x)
exp_x_negatve = np.exp(-x)
y = (exp_x_positve - exp_x_negatve) / (exp_x_positve + exp_x_negatve)

plt.plot(x, y)
````

### `Q-9:` Write a program implement Boradcasting Rule to check if two array can be added or not.
Given tuples of shapes.
```
shape of a- (3,2,2)
shape of b- (2,2)

check_broadcast(a, b) -> return Boolean (True if can broadcasted, False other wise.)
```

````python
def check_broadcast(arr1_shape, arr2_shape):
    # Start checking dimensions from the end
    arr1_index = len(arr1_shape)  - 1
    arr2_index = len(arr2_shape)  - 1

    while arr1_index > -1 and arr2_index > -1:
        arr1_dim = arr1_shape[arr1_index]
        arr2_dim = arr2_shape[arr2_index]
        # It should be either equal, or one of the dimensions should be 1
        if arr1_dim != arr2_dim and arr1_dim != 1 and arr2_dim != 1:
            return False
        arr1_index -= 1
        arr2_index -= 1
    # after comparing the elements returning true bcz 1 can be appended in the head always infinitely
    return True

a = (0, 3)
b = (3, )

print(check_broadcast(a, b))
````
